```python
# définition de lecture de requêtes SQL en python :

def requests_sql(requete, conn):
    """
    Exécute une requête SQL et retourne un DataFrame pandas.
    """
    return pd.read_sql(requete, conn)
```

```python
# Exercice N°11 : Pour chaque pays, affiche le nombre de clients et le total des ventes réalisées dans ce pays. 
# Inclue aussi les pays qui n'ont aucun achat (avec 0 pour le total des ventes)."

query = ''' 
SELECT 
    customers.Country, 
    COUNT(invoices.InvoiceId) AS NbInvoice, 
    COALESCE(SUM(invoices.Total), 0) AS TotalSales
FROM customers
LEFT JOIN invoices ON invoices.CustomerId = customers.CustomerId
GROUP BY customers.Country
ORDER BY TotalSales ASC
'''
df = requests_sql(query, conn)
df.head()
```

### Exercices du 4 Juillet 2025 :

```python
# Exercice N°12 : List all invoices including customer name, country, billing date, total amount, and employee (sales support) name who handled the invoice.

query = ''' 
SELECT 
    InvoiceId, InvoiceDate, Total, 
    customers.FirstName || ' ' || customers.LastName AS CustomerFullName, customers.Country,
    employees.FirstName || ' ' || employees.LastName AS EmployeeFullName
FROM invoices
JOIN customers ON invoices.CustomerId = customers.CustomerId
Join employees ON employees.EmployeeId = customers.SupportRepId
'''
df = requests_sql(query, conn)
df.head()
```

```python
# Exercice N°13 : List all customers whose total spend is above the average customer spend

query = ''' 
WITH TotalSpent AS (SELECT CustomerId, SUM(Total) AS TotalSells
FROM invoices
GROUP BY CustomerId)
SELECT FirstName, LastName, TotalSells
FROM customers
JOIN TotalSpent ON TotalSpent.CustomerId = customers.CustomerId
WHERE TotalSells > (SELECT AVG(TotalSells) FROM TotalSpent)
'''
df = requests_sql(query, conn)
df.head()
```

```python
# Exercice N°14 : For each artist, show number of tracks sold (invoice lines) and total sales amount. Only include artists with total sales > $100

query = ''' 
SELECT artists.name, COUNT(InvoiceLineId) AS TracksSold , SUM(invoice_items.UnitPrice) AS TotalSales
FROM artists
JOIN albums ON albums.ArtistId = artists.ArtistId
JOIN tracks ON tracks.AlbumId = albums.AlbumId
JOIN invoice_items ON invoice_items.TrackId = tracks.trackId
GROUP BY artists.name
HAVING TotalSales >= 100
'''
df = requests_sql(query, conn)
df.head()
```

```python
# Exercice N°15 : Show each billing country with its total sales and % of total global sales

query = ''' 
SELECT
    BillingCountry,
    SUM(Total) AS CountrySales,
    ROUND(100.0 * SUM(Total) / (SELECT SUM(Total) FROM invoices), 2) AS PercentOfGlobalSales
FROM invoices
GROUP BY BillingCountry
ORDER BY CountrySales DESC
'''
df = requests_sql(query, conn)
df.head()
```

```python
# Exercice N°16 : Rank customers by total spend, from highest to lowest

query = ''' 
SELECT FirstName || ' ' || LastName AS FullName, SUM(Total), RANK() OVER (ORDER BY SUM(Total) DESC) AS sales_rank
FROM customers
JOIN invoices ON invoices.CustomerId = customers.CustomerId
GROUP BY customers.FirstName, customers.LastName
'''
df = requests_sql(query, conn)
df.head()
```
